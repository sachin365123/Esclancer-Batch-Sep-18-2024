# Step 1: Upload HTML file here

```python
with open('/content/sample_data/home.html') as file:
  content=file.read()
```

# Step 2: Print content of this html file

```python
print(content)
```

# Step 3: Import Beautiful Libray

```python
from bs4 import BeautifulSoup
```

# Step 4: Parse content of HTML file

```python
new_soup= BeautifulSoup(content, "html.parser")
```

```python
print(new_soup)
```

# Step 5: Print only Head of HTML page

```python
print(new_soup.head)
```

# Step 6: Print only Body of HTML page

```python
print(new_soup.body)
```

# Step 7: Print only List of HTML page

```python
print(new_soup.li)
```

```python
print(new_soup.p)
```

```python
print(new_soup.h1)
```

```python
print(new_soup.li.string)
```

# Step 8: Print Hyperlinks of HTML page

```python
anchor_tags=new_soup.find_all(name='a')
print(anchor_tags)
```

# Step 9: Print all text from Hyperlinks of HTML page

```python
for tag in anchor_tags:
  print(tag.getText())
```

```python
anchor_tags=new_soup.find_all(name='p')
print(anchor_tags)
```

```python
for tag in anchor_tags:
  print(tag.getText())
```

```python
anchor_tags=new_soup.find_all(name='li')
print(anchor_tags)
```

```python
for tag in anchor_tags:
  print(tag.getText())
```

# Step 10: Parse multiple items from HTML page

```python
appmillers=new_soup.select_one(selector='p a')
print(appmillers)
```

```python
print(appmillers.get("href"))
```
